## 1\. Preços `outliers` de diamante

````
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
````

````
with sns.axes_style('whitegrid'):

  grafico = sns.boxplot(x=data["price"], y= data["cut"])
  grafico.set(title='Distribuição de Preço', xlabel='Preço',ylabel='cut' );
````

**Insight do gráfico 1**: Os pontos acima ou abaixo dos "whiskers" são potenciais outliers, ou seja, valores atípicos que estão significativamente mais altos ou mais baixos do que a maioria dos preços,  Isso pode indicar diamantes excepcionalmente valiosos. Os Diamantes de cortes "ideal" são os que
possuem mais outliers por sua caixa se encontra abaixo dos 5000 dolares, alguns considerados outliers se encontram acima 10000, a media fica abaixo dos 2500.
